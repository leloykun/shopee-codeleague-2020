```python
import pandas as pd
from functools import reduce
```

```python
def diff_cnt(s1, s2):
    return (s1 != s2).sum()
```

```python
pd_dfs = []
cat_cols = []
for i in range(1, 28+1):
    if i == 18: continue
    pd_df = pd.read_csv('submission_{}.csv'.format(i))
    pd_df['category'] = pd_df['category'].apply(lambda x : '{:02d}'.format(x))
    pd_df.columns     = ['filename', "category_{}".format(i)]
    cat_cols.append("category_{}".format(i))
    pd_dfs.append(pd_df)
    print("done with", i)

pd_merged = reduce(lambda  left, right: pd.merge(left, right, on=['filename'],
                                                 how='outer'), pd_dfs)
pd_merged.tail()
```

```python
for i in range(1, 28+1):
    if i == 18: continue
    print(i, diff_cnt(pd_merged['category_{}'.format(i)], pd_merged['category_25']))
```

```python
accepted_cats = [28, 26, 25, 13, 6, 2, 1]
accepted_cat_cols = ['category_{}'.format(cat) for cat in accepted_cats]
accepted_cat_cols
```

```python
pd_submission = pd_merged[['filename']].copy()
pd_submission['category'] = pd_merged[accepted_cat_cols].mode(axis=1)[0]
pd_submission.head()
```

```python
pd_submission.to_csv("submission_1_to_28_ensemble.csv", index=False)
```
