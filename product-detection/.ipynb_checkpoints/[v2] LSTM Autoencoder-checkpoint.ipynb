## Prepare Libraries and Global Variables

```python
import os
import _pickle as cPickle

import numpy as np
import pandas as pd
import datetime

import statsmodels.api as sm
from sklearn.preprocessing import MinMaxScaler, LabelEncoder
from sklearn.ensemble import IsolationForest

from densmap import densMAP 
import umap.umap_ as umap
import umap.plot

import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

import tensorflow.keras as tfk
import tensorflow as tf
```

```python
FORECAST_HORIZON = 30
ROLLING_WINDOW_SKIP = 1
TS_LEN_LIM = 300         # limit length of each timeseries
TS_LIM = 50

DIR = "datasets/clean/"

EPS = 1e-9
```

## Load Dataset

```python
df_sources = pd.read_csv('dataset_sources_train.csv')
```

```python
%%time
data_blocks     = []
data_origins    = []
data_ts_ids     = []
data_freqs      = []

for origin in np.unique(df_sources['origin']):
    print(origin)
    i = np.where(df_sources['origin'] == origin)
    
    df_sample = df_sources.iloc[i].sample(frac=1).iloc[:TS_LIM]
    for id, row in df_sample.iterrows():
        origin = row['origin']
        freq   = row['freq']
        source = row['source']

        df_data = pd.read_csv(source)
        
        #original_min_y = df_data['y'].min()
        #original_max_y = df_data['y'].max()
        #df_data['y'] = (df_data['y'] - original_min_y) / (original_max_y - original_min_y)
        
        START = max(0, len(df_data)-FORECAST_HORIZON-TS_LEN_LIM)
        for i in range(START, len(df_data)-FORECAST_HORIZON, ROLLING_WINDOW_SKIP):
            data_blocks.append(df_data['resid_scaled'][i:i+FORECAST_HORIZON])
            data_origins.append(origin)
            data_ts_ids.append(id)
            data_freqs.append(freq)
```

```python
data_blocks     = np.array(data_blocks).reshape(-1, FORECAST_HORIZON, 1)
data_origins    = np.array(data_origins)
data_ts_ids     = np.array(data_ts_ids)
data_freqs      = np.array(data_freqs)
```

## Prepare LSTM Autoencoder

```python
original_dim = (FORECAST_HORIZON, 1)
latent_dim = 4
```

```python
encoder = tfk.models.Sequential([
    tfk.layers.Input(original_dim),
    #tfk.layers.LSTM(4**3*latent_dim, activation='relu', return_sequences=True, name='encoder_lstm_1'),
    tfk.layers.LSTM(4**2*latent_dim, activation='relu', return_sequences=True, name='encoder_lstm_2'),
    tfk.layers.LSTM(4*latent_dim, activation='relu', return_sequences=True, name='encoder_lstm_3'),
    tfk.layers.LSTM(latent_dim, activation='tanh', name='encoder_lstm_4'),
], name='encoder')
encoder.summary()
```

```python
decoder = tfk.models.Sequential([
    tfk.layers.RepeatVector(original_dim[0], input_shape=(latent_dim,), name='decoder_repeat_vector'),
    tfk.layers.LSTM(4*latent_dim, activation='relu', return_sequences=True, name='decoder_lstm_1'),
    tfk.layers.LSTM(4**2*latent_dim, activation='relu', return_sequences=True, name='decoder_lstm_2'),
    #tfk.layers.LSTM(4**3*latent_dim, activation='relu', return_sequences=True, name='decoder_lstm_3'),
    tfk.layers.TimeDistributed(tfk.layers.Dense(original_dim[1], activation='sigmoid'), name='decoder_time_distributed_dense')
], name='decoder')
decoder.summary()
```

```python
model = tfk.models.Sequential([
    tfk.layers.Input(original_dim),
    encoder,
    decoder
], name='autoencoder')


model.compile(optimizer=tfk.optimizers.Adam(learning_rate=1e-3), 
              loss='mse')

model.summary()
```

## Train LSTM Autoencoder

```python
reduce_lr = tf.keras.callbacks.ReduceLROnPlateau(
    monitor='val_loss',
    factor=0.5,
    patience=3,
    min_lr=1e-5
)

early_stop = tfk.callbacks.EarlyStopping(
    monitor='val_loss',
    patience=10,
    restore_best_weights=True
)

checkpoint_dir = os.path.join('.model_files')
if not os.path.exists(checkpoint_dir):
    os.makedirs(checkpoint_dir)
checkpoint = tfk.callbacks.ModelCheckpoint(
    filepath = checkpoint_dir + '/ckpt/',
    save_freq ='epoch',
    save_best_only=True
)

log_dir = os.path.join(".model_files/logs/" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S"))
if not os.path.exists(log_dir):
    os.makedirs(log_dir)
tensorboard_callback = tf.keras.callbacks.TensorBoard(
    log_dir=log_dir,
    histogram_freq=1,
    update_freq=100,
    profile_batch = 100000000
)
```

```python
hist = model.fit(data_blocks, data_blocks,
                 epochs=1000, batch_size=256,
                 validation_split=0.2,
                 callbacks=[reduce_lr,
                            early_stop, 
                            #checkpoint, 
                            tensorboard_callback
                           ])
```

```python
pd.DataFrame(hist.history).plot()
```
